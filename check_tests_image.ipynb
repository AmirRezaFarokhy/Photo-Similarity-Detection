```python
import pandas as pd
import matplotlib.pyplot as plt
import cv2 

from ClockDetection import FindSimilar

# Test One
labels = pd.read_csv("labels.csv")
file_path = "test/test_1.jpg"
SHAPES = (220, 420)

finder = FindSimilar(SHAPES, file_path)
accuracy = finder.CompareImages(labels)
maximom = max(accuracy)[0]
for val in accuracy:
    if val[0]==maximom and maximom!=0:
        print(val)
```

```python
maximom = max(accuracy)[0]
high_accuracy = []
for val in accuracy:
    if val[0]==maximom and maximom!=0:
        high_accuracy.append(val[1].split('/')[1])

for index in range(len(labels)-1):
    if labels['path'].iloc[index]==high_accuracy[0]:
        print(f"We find the prediction time nears {labels['target'].iloc[index][1]} hour : {labels['target'].iloc[index][-3:]} minutes")
        
        
fig = plt.figure(figsize=(10, 14))
rows, cols = 2, 1

fig.add_subplot(rows, cols, 1)  
img_accuracy = cv2.imread(f"created_dataset/{high_accuracy[-1]}")
plt.imshow(img_accuracy)
plt.title("Prediction Image")

fig.add_subplot(rows, cols, 2)
plt.imshow(cv2.imread(file_path, cv2.COLOR_BGR2GRAY))
plt.title("Real Image")

plt.show()
```
